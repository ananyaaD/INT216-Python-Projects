## Calculator using tkinter library 

```python
from tkinter import *

window = Tk()

window.geometry("312x324")

window.resizable(0, 0)

window.title("Calculator")


#update the input field whenever a number is entered or any button is pressed
def btn_click(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)

#clears the input field or previous calculations using C button
def btn_clear():
    global expression
    expression = ""
    input_text.set("")

#to calculate the expression in input field and show the output
def btn_equal():
    try:
        global expression
        result = str(eval(expression))
        input_text.set(result)
        expression = ""
    except:
        input_text.set('ERROR')
        expression=''

#remove the last entered digit/character   
def btn_back():
    global expression
    expression = expression[:-1]
    input_text.set(expression)

expression = ""
input_text = StringVar()


input_frame = Frame(window, width = 312, height = 50, bd = 0, highlightcolor = "black", highlightthickness = 1)
input_frame.pack(side = TOP)


input_field = Entry(input_frame, font = ('arial', 18, 'bold'), textvariable = input_text, width = 50, bg = "white", bd = 0, justify = RIGHT)
input_field.grid(row = 0, column = 0)
input_field.pack(ipady = 10) 


btns_frame = Frame(window, width = 312, height = 272.5, bg = "grey")
btns_frame.pack()

#===================================================BUTTON WIDGETS======================================================================

clear = Button(btns_frame, text = "C", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro",  command = lambda: btn_clear()).grid(row = 0, column = 0, padx = 1, pady = 1)
back = Button(btns_frame, text = "âŒ«", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro",  command = lambda: btn_back()).grid(row = 0, column = 1, padx = 1, pady = 1)
mod = Button(btns_frame, text = "mod", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro",  command = lambda: btn_click("%")).grid(row = 0, column = 2, padx = 1, pady = 1)
divide = Button(btns_frame, text = "/", fg = "black", width = 10, height = 3, bd = 0,bg = "gainsboro",  command = lambda: btn_click("/")).grid(row = 0, column = 3, padx = 1, pady = 1)

seven = Button(btns_frame, text = "7", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(7)).grid(row = 1, column = 0, padx = 1, pady = 1)
eight = Button(btns_frame, text = "8", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(8)).grid(row = 1, column = 1, padx = 1, pady = 1)
nine = Button(btns_frame, text = "9", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(9)).grid(row = 1, column = 2, padx = 1, pady = 1)
multiply = Button(btns_frame, text = "*", fg = "black", width = 10, height = 3, bd = 0,bg = "gainsboro", command = lambda: btn_click("*")).grid(row = 1, column = 3, padx = 1, pady = 1)


four = Button(btns_frame, text = "4", fg = "black", width = 10, height = 3, bd = 0,  cursor = "hand2", command = lambda: btn_click(4)).grid(row = 2, column = 0, padx = 1, pady = 1)
five = Button(btns_frame, text = "5", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(5)).grid(row = 2, column = 1, padx = 1, pady = 1)
six = Button(btns_frame, text = "6", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(6)).grid(row = 2, column = 2, padx = 1, pady = 1)
minus = Button(btns_frame, text = "-", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro", command = lambda: btn_click("-")).grid(row = 2, column = 3, padx = 1, pady = 1)


one = Button(btns_frame, text = "1", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(1)).grid(row = 3, column = 0, padx = 1, pady = 1)
two = Button(btns_frame, text = "2", fg = "black", width = 10, height = 3, bd = 0,  cursor = "hand2", command = lambda: btn_click(2)).grid(row = 3, column = 1, padx = 1, pady = 1)
three = Button(btns_frame, text = "3", fg = "black", width = 10, height = 3, bd = 0, cursor = "hand2", command = lambda: btn_click(3)).grid(row = 3, column = 2, padx = 1, pady = 1)
plus = Button(btns_frame, text = "+", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro", command = lambda: btn_click("+")).grid(row = 3, column = 3, padx = 1, pady = 1)


zero = Button(btns_frame, text = "0", fg = "black", width = 21, height = 3, bd = 0,  cursor = "hand2", command = lambda: btn_click(0)).grid(row = 4, column = 0, columnspan = 2, padx = 1, pady = 1)
point = Button(btns_frame, text = ".", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro",  command = lambda: btn_click(".")).grid(row = 4, column = 2, padx = 1, pady = 1)
equals = Button(btns_frame, text = "=", fg = "black", width = 10, height = 3, bd = 0, bg = "gainsboro",  command = lambda: btn_equal()).grid(row = 4, column = 3, padx = 1, pady = 1)



window.mainloop()
```
